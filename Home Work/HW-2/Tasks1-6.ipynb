## 3. Верно ли, что студенты, имеющие больше 10 пропусков, учатся хуже, чем остальные?

```python
# mt - more that 10
#lt - less than 10
mt = df[df.absences > 10]
lt = df[df.absences <= 10]
mt_grade = mt['G3'].mean()
lt_grade = lt['G3'].mean()
print(mt.shape, lt.shape)
print('Средняя оценка тех, кто проугливал больше 10 раз:', mt_grade)
print('Средняя оценка тех, кто прогуливал менее 10 раз:', lt_grade)
```

#### Вывод: 
Верно, студенты, с 10-ью пропусками и больше учатся хуже, однако не на много, всего на 0.3 балла. Так же стоит принять во внимание то, что кол-во редко прогуливающих студентов в 6 раз больше.

## 4. Как вы думаете, какие еще признаки могут повлиять на итоговую оценку? Найдите закономерности, аналогичные рассмотренным.

- ### Желание высшего образования

```python
want = df[df.higher == 'yes']
doesnt_want = df[df.higher == 'no']
print(want.shape, doesnt_want.shape)
print('Средний балл желающих получить ВО', want['G3'].mean())
print('Средний балл не желающих получить ВО', doesnt_want['G3'].mean())
```

#### Вывод:
Разница в средних баллах значительна: 10.6 против 6.8. <br>
Можно считать, что желание высшего образования является мотивацией к стремлению к высоким баллам, однако опять же кол-во студентов в выборках сильно разнятся: 375 - желающих получить ВО, 20 - нежелающих.

- ### В отношениях студент или нет

```python
related = df[df.romantic == 'yes']
not_related = df[df.romantic == 'no']
print(related.shape, not_related.shape)
print('Средний балл людей в отношениях', related['G3'].mean())
print('Средний балл одиноких людей', not_related['G3'].mean())
```

#### Вывод:
Разница в средних баллах присутствует: 9.58 против 10.84 <br>
Отношения влияют на успехи в учёбе. У тех, кто встречается, баллы меньше чем у людей без пары.

- ### Ходит ли студент на платные занятия

```python
goes = df[df.paid == 'yes']
doesnt_goes = df[df.paid == 'no']
print(goes.shape, doesnt_goes.shape)
print('Средний балл студентов, посещяющих платные занятия', goes['G3'].mean())
print('Средний балл студентов, не посещяющих платные занятия', doesnt_goes['G3'].mean())
```

#### Вывод:

Доп занятия помогают студентам получать более высокий балл

## 5. Как вы думаете, если при школе откроется общежитие, насколько актуально это будет для студентов математического класса?

Чтобы понять, актуальность открытия общежития, нужно узнать, как много студентов нуждаются в общежитии.<br> Студенты, которым нужно место в общаге:
- Живут в сельской местости
- Живут в городе, но далеко ехать
- Те, у кого большая семья, которая, возможно, отвлекает от учёбы

```python
need_studs = df[(df.address == 'R') | (df.traveltime >= 3) | (df.famsize == 'GT3')]
all_studs = df
len_need = len(need_studs)
len_all = len(all_studs)
print('Кол-во студентов нуждающихся в общежитии:', len_need)
print('Кол-во всех студентов:', len_all)
```

#### Вывод:
Открытие общежития актуально, так как 77 процентов студентов нуждаются в нём.

## 6. Найдите в данных закономерности с помощью группировки.

- ### Зависимость средней оценки от кол-ва часов обучения в неделю

```python
info = df.groupby('studytime').G3.mean()
x = info.index
y = info.values
fig, ax = plt.subplots()
ax.bar(x, y)
ax.grid()
plt.xlabel('Кол-во часов в неделю', fontsize=15)
plt.ylabel('Средняя итоговая оценка', fontsize=15)
plt.show()
info
```

#### Вывод:
Чем больше часов учится студент, тем выше его оценка. Однако если студент учится слишком много, то его баллы немного падают в виду переутомления.

- ### Зависимость средней оценки от состояния здоровья студента

```python
info = df.groupby('health').G3.mean()
x = info.index
y = info.values
fig, ax = plt.subplots()
ax.bar(x, y)
ax.grid()
plt.xlabel('Здоровье студента', fontsize=15)
plt.ylabel('Средняя итоговая оценка', fontsize=15)
plt.show()
```

#### Вывод:
У студентов, которые учатся лучше всех, здоровье - плохое. Вероятно они перенапрягаются, переутомляются ради достижения хорошего балла

- ### Зависимость средней оценки от отношений в семье

```python
info = df.groupby('famrel').G3.mean()
x = info.index
y = info.values
fig, ax = plt.subplots()
ax.bar(x, y)
ax.grid()
plt.xlabel('Отношения в семье', fontsize=15)
plt.ylabel('Средняя итоговая оценка', fontsize=15)
plt.show()
info
```

#### Вывод:
Студенты, у которых в семье идеальные отношения, либо ужасные, в среднем учатся лучше остальных
