$$\tilde{A}=\begin{pmatrix}

7 & 8 & 9
\end{matrix}\right|
\begin{matrix}
a\\

Найти соотношение между параметрами $a$, $b$ и $c$, при которых система является несовместной.

$$
\begin{pmatrix}
\left.\begin{matrix}
1 & 2 & 3\\ 
4 & 5 & 6\\ 
7 & 8 & 9
\end{matrix}\right|
\begin{matrix}
a\\ 
b\\
c
\end{matrix}
\end{pmatrix}
\Leftrightarrow
\begin{pmatrix}
\left.\begin{matrix}
1 & 2 & 3\\ 
3 & 3 & 3\\ 
7 & 8 & 9
\end{matrix}\right|
\begin{matrix}
a\\ 
b-a\\
c
\end{matrix}
\end{pmatrix}
\Leftrightarrow
\begin{pmatrix}
\left.\begin{matrix}
1 & 2 & 3\\ 
3 & 3 & 3\\ 
6 & 6 & 6
\end{matrix}\right|
\begin{matrix}
a\\ 
b-a\\
c-a
\end{matrix}
\end{pmatrix}
\Leftrightarrow
\begin{pmatrix}
\left.\begin{matrix}
1 & 2 & 3\\ 
3 & 3 & 3\\ 
0 & 0 & 0
\end{matrix}\right|
\begin{matrix}
a\\ 
b-a\\
(c-a) -2(b-a)
\end{matrix}
\end{pmatrix}
$$

Если не выполняется равенство $(c-a)-2(b-a) = c+a -2b = 0$, то СЛУ не является совместной

$$
